```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv("../data/2022.csv")
df
```

```python
df.rename(columns=lambda x: x.title(), inplace=True)
df['Program'] = df['Program'].str.title()
```

```python
df.columns[2:]
```

```python
for district in df.columns[2:]:
    df.loc[df[district] == "NQC", district] = 0.0
    df[district] = df[district].astype(float)
```

```python
df.to_csv('../data/2022_clean.csv', index=False)

print("DataFrame saved as CSV successfully.")
```
